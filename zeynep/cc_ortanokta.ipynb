```python
import pydicom
import numpy as np
import cv2
import matplotlib.pyplot as plt

# RCC ve LCC görüntülerin dosya yollarını belirtin
rcc_file_path = "C:/Users/Zeynep Aygün/Desktop/mamografi/images/hasta_no/RCC.dcm"
lcc_file_path = "C:/Users/Zeynep Aygün/Desktop/mamografi/images/hasta_no/LCC.dcm"

# RCC görüntüsünü okuyun
rcc_dcm = pydicom.read_file(rcc_file_path)
rcc_image = rcc_dcm.pixel_array

# LCC görüntüsünü okuyun
lcc_dcm = pydicom.read_file(lcc_file_path)
lcc_image = lcc_dcm.pixel_array

# RCC ve LCC görüntülerini BGR renk formatına dönüştürün
rcc_image_bgr = cv2.cvtColor(rcc_image, cv2.COLOR_GRAY2BGR)
lcc_image_bgr = cv2.cvtColor(lcc_image, cv2.COLOR_GRAY2BGR)

# RCC görüntüsündeki meme orta noktasını bulun
rcc_moments = cv2.moments(rcc_image)
rcc_cx = int(rcc_moments["m10"] / rcc_moments["m00"])
rcc_cy = int(rcc_moments["m01"] / rcc_moments["m00"])

# RCC görüntüsündeki meme orta noktasını işaretleyin
cv2.circle(rcc_image_bgr, (rcc_cx, rcc_cy), 5, (0, 0, 255), -1)

# LCC görüntüsündeki meme orta noktasını bulun
lcc_moments = cv2.moments(lcc_image)
lcc_cx = int(lcc_moments["m10"] / lcc_moments["m00"])
lcc_cy = int(lcc_moments["m01"] / lcc_moments["m00"])

# LCC görüntüsündeki meme orta noktasını işaretleyin
cv2.circle(lcc_image_bgr, (lcc_cx, lcc_cy), 5, (0, 0, 255), -1)

# RCC ve LCC görüntülerini gösterin
plt.imshow("RCC", rcc_image_bgr)
plt.imshow("LCC", lcc_image_bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def find_white_pixels(gray_img):
    # Beyaz piksellerin konumlarını bulma
    white_pixels = np.where(gray_img == 255)

    # Pikselleri listeye ekleme
    pixel_list = list(zip(white_pixels[0], white_pixels[1]))

    # Toplu olarak duran beyaz piksellerin en sağ ortasını bulma
    if len(pixel_list) > 0:
        rightmost_pixel = max(pixel_list, key=lambda x: x[1])
        leftmost_pixel = min(pixel_list, key=lambda x: x[1])
        y_mean = int((rightmost_pixel[1] + leftmost_pixel[1]) /2)
        x_mean = int(np.mean([pixel[0] for pixel in pixel_list]))
        center_point = (y_mean, x_mean)
        return pixel_list, center_point
    else:
        return [], None
```

```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt

# .nii.gz formatındaki maskeyi yükleyin
mask_nii = nib.load("C:/Users/Zeynep Aygün/Desktop/mamografi/masks/hasta_no/RCC.nii.gz")
mask_data = mask_nii.get_fdata()

# Beyaz piksellerin konumlarını bulun
white_pixels = np.argwhere(mask_data == 1)

# Beyaz piksellerin x, y, z koordinatlarını ayrı listelerde saklayın
x_coords = white_pixels[:, 0]
y_coords = white_pixels[:, 1]
z_coords = white_pixels[:, 2]

# Beyaz piksellerin orta noktasını bulun
center_x = int(np.mean(x_coords))
center_y = int(np.mean(y_coords))
center_z = int(np.mean(z_coords))

# Orta noktayı işaretleyin
plt.imshow(mask_data[:, :, center_z], cmap='gray')  # Sadece orta düzlemi göster
plt.scatter(center_y, center_x, c='red', s=10)
plt.title('Orta Nokta')
plt.show()

# Belirli bir pixel kaydırma yaparak istenen miktarda sağa taşıyın
pixel_shift = 10  # Kaydırma miktarı
new_center_y = center_y + pixel_shift

# Yeni orta noktayı işaretleyin
plt.imshow(mask_data[:, :, center_z], cmap='gray')
plt.scatter(new_center_y, center_x, c='green', s=10)  # Yeni orta noktayı yeşil renkte işaretle
plt.title('Yeni Orta Nokta')
plt.show()
```
