```python
import pydicom
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt

def crop_and_mask_dcm_image(dcm_path, mask_path):
    # Oku .dcm ve mask görüntülerini
    imgs = pydicom.dcmread(dcm_path).pixel_array
    masks = nib.load(mask_path).get_fdata()

    # .dcm görüntüsünü ters çevir, normalize et ve uint8 tipine dönüştür
    imgs=np.invert(imgs)
    imgs = ((imgs - imgs.min()) * (1/(imgs.max() - imgs.min()) * 255)).astype('uint8')

    # Maskeyi uygun şekilde düzenle
    if 'Sinem' in dcm_path :
        if 'LCC' in dcm_path:
            masks=masks[:,:,1]
        elif 'LMLO' in dcm_path:
            masks=masks[:,:,3]
        elif 'RCC' in dcm_path:
            masks=masks[:,:,0]
        elif 'RMLO' in dcm_path:
            masks=masks[:,:,2]
    elif 'Ertan' in dcm_path:
        masks=masks[:,:,0]

    elif 'Zeynep' in dcm_path:
        if 'LCC' in dcm_path:
            masks=masks[:,:,0]
        elif 'LMLO' in dcm_path:
            masks=masks[:,:,1]
        elif 'RCC' in dcm_path:
            masks=masks[:,:,2]
        elif 'RMLO' in dcm_path:
            masks=masks[:,:,3]
            
    masks[masks==1]=0
    masks[masks==3]=1
    masks[masks==2]=1
    
    masks = np.flip(masks)
    masks = np.rot90(masks)
    masks = np.fliplr(masks)

    # Maskelenmiş görüntüyü hesapla
    rslt = masks * imgs

    # Maske içindeki alanı sınırlayacak bounding box'i hesapla
    rows, cols = np.where(masks != 0)
    y1, x1 = np.min(rows), np.min(cols)
    y2, x2 = np.max(rows), np.max(cols)

    # Görüntüyü bounding box ile kırp
    result = rslt[y1:y2+1, x1:x2+1]

    # Maske dışındaki alanı siyah renge dök
    result[masks[y1:y2+1, x1:x2+1] == 0] = 0

    return imgs, rslt, result

# Örnek kullanım
dcm_path = "C:/Users/Zeynep Aygün/Desktop/mamografi/images/hasta_no/RMLO.dcm"
mask_path = "C:/Users/Zeynep Aygün/Desktop/mamografi/masks/hasta_no/RMLO.nii.gz"

imgs, rslt, result = crop_and_mask_dcm_image(dcm_path, mask_path)

# Sonuçları görüntüle
plt.subplot(1, 3, 1)
plt.imshow(imgs, cmap="gray")
plt.title("Orjinal .dcm Görüntüsü")
plt.subplot(1, 3, 2)
plt.imshow(rslt, cmap="gray")
plt.title("Kesilmiş Görüntü")
plt.subplot(1, 3, 3)
plt.imshow(result, cmap="gray")
plt.title("Maskenin Beyaz Alanlarını Kesme ve İnvert Etme")
plt.show()
```
