```python
import os 
output_path='/home/uraninjo/Desktop/Çalışmalar/Mergen1-Teknofest/Murat/BIRADSPosta1'
try:
    os.mkdir(output_path)
except:
    print('directory already exists')
os.mkdir(f'{output_path}/data')
for i in df['BIRADS KATEGORİSİ'].unique():
    os.mkdir(f'{output_path}/data/{i}')
```

```python
input_path='/home/uraninjo/Desktop/Çalışmalar/Mergen1-Teknofest/Murat/TestPipeline/Teknofest_Version1_Segmentation_Output'
l=glob.glob('/home/uraninjo/Desktop/Çalışmalar/Mergen1-Teknofest/Murat/TestPipeline/Teknofest_Version1_Segmentation_Output/*')
for path in l:
    hno=path.split('/')[-1]
    birads=df['BIRADS KATEGORİSİ'][df.HASTANO==int(hno)].item()
    durumdf=df[df.HASTANO==int(hno)]
    for lr in ['KADRAN BİLGİSİ (SAĞ)','KADRAN BİLGİSİ (SOL)']:
        if not durumdf[lr].item() is np.nan and 'SAĞ' in lr:
            try:
                shutil.copy(f'{path}/RMLO.png',f'{output_path}/data/{birads}/{hno}.png')
                shutil.copy(f'{path}/RCC.png',f'{output_path}/data/{birads}/{hno}-2.png')
            except Exception as e:
                print(e)
        if not durumdf[lr].item() is np.nan and 'SOL' in lr:
            try:
                shutil.copy(f'{path}/LMLO.png',f'{output_path}/data/{birads}/{hno}-3.png')
                shutil.copy(f'{path}/LCC.png',f'{output_path}/data/{birads}/{hno}-4.png')
            except Exception as e:
                print(e)


        if durumdf[lr].item() is np.nan and 'SAĞ' in lr:
            try:
                shutil.copy(f'{path}/RMLO.png',f'{output_path}/data/BI-RADS1-2/{hno}-5.png')
                shutil.copy(f'{path}/RCC.png',f'{output_path}/data/BI-RADS1-2/{hno}-6.png')
            except Exception as e:
                print(e)
        if durumdf[lr].item() is np.nan and 'SOL' in lr:
            try:
                shutil.copy(f'{path}/LMLO.png',f'{output_path}/data/BI-RADS1-2/{hno}-7.png')
                shutil.copy(f'{path}/LCC.png',f'{output_path}/data/BI-RADS1-2/{hno}-8.png')
            except Exception as e:
                print(e)
```
